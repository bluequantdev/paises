```python
import requests
import yfinance as yf
import pandas as pd

url = "https://api.bitget.com/api/spot/v1/market/tickers"
res = requests.get(url)
bitget_data = res.json().get("data", [])

bitget_paises = {}
for item in bitget_data:
    symbol = item.get("symbol", "")
    volume = float(item.get("baseVolume", 0))
    moeda = symbol[-4:]
    if moeda == "USDT":
        pais = "Global (Cripto)"
        bitget_paises[pais] = bitget_paises.get(pais, 0) + volume

acoes = ["AAPL", "MSFT", "NVDA", "GOOGL", "AMZN", "JNJ", "TSM", "BABA", "ITUB", "VALE", "RY", "TD", "INFY"]

dados_acoes = []
for t in acoes:
    try:
        tk = yf.Ticker(t)
        info = tk.info
        dados_acoes.append({
            "Ticker": t,
            "Country": info.get("country", "N/A"),
            "Market Cap": info.get("marketCap", 0)
        })
    except:
        continue

df_acoes = pd.DataFrame(dados_acoes)
df_agrupado = df_acoes.groupby("Country").sum().reset_index()

for pais, volume in bitget_paises.items():
    nova_linha = pd.DataFrame([{"Country": pais, "Market Cap": volume * 1_000_000}])
    df_agrupado = pd.concat([df_agrupado, nova_linha], ignore_index=True)

df_agrupado.rename(columns={"Market Cap": "Total"}, inplace=True)
df_agrupado.to_csv("alocacao_por_pais.csv", index=False)
```
